2Ô∏è‚É£ Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

3Ô∏è‚É£ Load Dataset

```python
import pandas as pd

df = pd.read_csv("data/Reviews.csv")
df.head()
```

4Ô∏è‚É£ Rating Distribution Visualization
üîπ Bar Chart: Rating Frequency

```python
sns.countplot(x="Score", data=df)
plt.title("Distribution of Customer Ratings")
plt.xlabel("Rating Score")
plt.ylabel("Number of Reviews")
plt.show()
```

üìå Insight:

Most reviews are concentrated around higher ratings (4‚Äì5).

Indicates overall positive customer satisfaction.

5Ô∏è‚É£ Review Length Analysis
üîπ Histogram: Review Text Length

```python
df["review_length"] = df["Text"].astype(str).apply(len)

plt.hist(df["review_length"], bins=50)
plt.title("Distribution of Review Lengths")
plt.xlabel("Number of Characters")
plt.ylabel("Frequency")
plt.show()
```

üìå Insight:

Majority of reviews are short to medium length.

A smaller number of detailed reviews exist, often reflecting strong opinions.

6Ô∏è‚É£ Helpfulness Votes vs Rating
üîπ Scatter Plot

```python
sns.scatterplot(x="Score", y="HelpfulnessNumerator", data=df)
plt.title("Rating vs Helpfulness Votes")
plt.xlabel("Rating Score")
plt.ylabel("Helpful Votes")
plt.show()
```

üìå Insight:

Reviews with extreme ratings (1 or 5) tend to receive more helpful votes.

Strong opinions attract more user engagement.

7Ô∏è‚É£ Reviews Over Time
üîπ Line Plot: Review Trends

```python
df["Time"] = pd.to_datetime(df["Time"], unit="s")
df["Year"] = df["Time"].dt.year

reviews_per_year = df["Year"].value_counts().sort_index()

plt.plot(reviews_per_year.index, reviews_per_year.values)
plt.title("Number of Reviews Over Time")
plt.xlabel("Year")
plt.ylabel("Review Count")
plt.show()
```

üìå Insight:

Review volume increases over time.

Reflects growing adoption of online reviews and Amazon platform usage.

8Ô∏è‚É£ Sentiment-Based Visualization (Optional but Strong)

If sentiment labels exist or were derived earlier:

üîπ Sentiment Distribution

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from textblob import TextBlob

# Load dataset
df = pd.read_csv("data/Reviews.csv")

# Drop missing text
df = df.dropna(subset=["Text"])

# TAKE SAMPLE (VERY IMPORTANT)
df = df.sample(5000, random_state=42)

# Create Sentiment column
def get_sentiment(text):
    polarity = TextBlob(str(text)).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment"] = df["Text"].apply(get_sentiment)

print(df["Sentiment"].value_counts())

# Plot
sns.countplot(
    x="Sentiment",
    data=df,
    order=["Positive", "Neutral", "Negative"]
)
plt.title("Sentiment Distribution of Reviews (Sampled Data)")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

üìå Insight:

Positive sentiment dominates the dataset.

Confirms alignment between numerical ratings and textual sentiment.
