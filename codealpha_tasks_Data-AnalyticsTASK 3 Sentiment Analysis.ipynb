5Ô∏è‚É£ Sentiment Polarity Calculation

Using TextBlob, sentiment polarity is calculated for each review:

Polarity range: ‚Äì1 to +1

Negative ‚Üí Neutral ‚Üí Positive

```python
from textblob import TextBlob
import pandas as pd

# Reload dataset (safe)
df = pd.read_csv("data/Reviews.csv")

# Drop missing text
df = df.dropna(subset=["Text"])

# TAKE SAMPLE to avoid freezing
df = df.sample(5000, random_state=42)

# Polarity function
def get_polarity(text):
    return TextBlob(str(text)).sentiment.polarity

# Apply polarity
df["Polarity"] = df["Text"].apply(get_polarity)

# Check
df[["Polarity"]].head()
```

```python
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

df['Polarity'] = df['Text'].apply(get_polarity)
```

6Ô∏è‚É£ Sentiment Classification

```python
def get_sentiment(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df['Sentiment'] = df['Polarity'].apply(get_sentiment)
```

This classifies each review into:

Positive

Negative

Neutral

7Ô∏è‚É£ Sentiment Distribution Visualization

```python
sns.countplot(x="Sentiment", data=df)
plt.title("Sentiment Distribution of Amazon Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
üìå Insight:

Majority of reviews fall under Positive sentiment

Indicates high overall customer satisfaction

Neutral reviews represent factual or balanced feedback
```

8Ô∏è‚É£ Sentiment vs Rating Analysis

```python
sns.boxplot(x="Sentiment", y="Score", data=df)
plt.title("Rating Distribution Across Sentiments")
plt.show()
```

üìå Insight:
